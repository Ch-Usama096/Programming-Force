```python
import cv2
import numpy as np
import pandas as pd
from PIL import Image
import os
from tensorflow.keras.preprocessing import image 
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import ResNet152V2 , InceptionV3 , Xception # Import the Resnet 50 Model
from keras.layers import GlobalAvgPool2D as GAP, Dense, Dropout
from tensorflow.keras.applications.resnet50 import preprocess_input
```

<b>Load the Resnet Pre-Trained Model

```python
model_h5_path = "E:\\software\\ResNet.h5"
resnetModel = load_model(model_h5_path)
```

```python
# Display the Summary of the ResNet Model
resnetModel.summary()
```

<b>Visulization the Prediction Images

```python
img_path = "E:\\software\\shoe_boot_sandal"

plt.figure(figsize = (10,10))
i = 1
for images in os.listdir(img_path):
    # Get the Path of the Each Image
    path_image = os.path.join(img_path , images)
   
    # Read the Image
    image = cv2.imread(path_image)
    
    image = cv2.resize(image , (256,256))
    
    plt.subplot(4,5,i)
    plt.imshow(image)
    plt.xticks([])
    plt.yticks([])
    
    i += 1
    
    if i >= 21:break
plt.show()
```

<b>Predict the Result with Resnet PreTrained Model

```python
# Define the Class Label
class_label = ['Boot', 'Sandal', 'Shoe']
```

```python
count = 1
for images in os.listdir(img_path):
    # Get the Path of the Each Image
    path_image = os.path.join(img_path , images)
   
    # Read the Image
    image = cv2.imread(path_image)
    image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
    
    # Resize the Image
    resize_image = cv2.resize(image , (256,256))
    
    img=np.expand_dims(resize_image,axis=0)
    
    # Now Predict the Result with ResNet PreTrained Model
    res = np.argmax(resnetModel.predict(img))
    
    label = class_label[int(res)]
    
    # Display the Image with Label (If You want to, separate the image you just apply the condtion and send the image with condition base folder)
    
    print(label)
    # Display the Image
    display(Image.fromarray(image))
```
