```python
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

<b>Load & Display the Image

```python
image = cv2.imread(r"C:\Users\User131022\Desktop\MonkeyPF\monkey.png")

# Image
display(Image.fromarray(cv2.cvtColor(image , cv2.COLOR_BGR2RGB)))
```

<b>Detect the Red Color and Replace Black Color

```python
# Convert the Image into HSV 
imageHSV = cv2.cvtColor(image , cv2.COLOR_BGR2HSV)    


# Define the Lower & Upper Value of Red Color
lowerValue  = np.array([0, 100, 100])
HigherValue = np.array([9,255,255])


# Creating the Mask of red Color
maskRed = cv2.inRange(imageHSV , lowerValue , HigherValue)

# Convert the Red Color to Black Color
resImage = cv2.cvtColor(imageHSV , cv2.COLOR_HSV2BGR)
resImage[maskRed > 0] = 0


# Display  All Images
imageTitle = ['Actaul Image' , 'HSV Image' , 'Mask Image' , 'Result Image']
images     = [cv2.cvtColor(image , cv2.COLOR_BGR2RGB) , imageHSV , maskRed , cv2.cvtColor(resImage , cv2.COLOR_BGR2RGB)]

plt.figure(figsize = (10,10))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(imageTitle[i])
    plt.xticks([]) , plt.yticks([])
plt.show()
```
