```python
import numpy as np
import cv2
from PIL import Image
import json
import os
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
# Load the Json File Data

with open("cat_dog_annotation.json" , "r") as f:
    json_data = json.load(f)
    
# Get the Folder Path
folder_path = "E:\\software\\Task2\\cat_dog\\"


# Extract the Infromation from Json File
for image_id , image_info in json_data["_via_img_metadata"].items():
    # Get the File Name of the Image
    image_name = image_info["filename"]
    
    # Now Get the Image
    img_path = os.path.join(folder_path , image_name)
    
    # Read the Image with cv2 
    image = cv2.imread(img_path)
    image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
    
     # Create figure and axes
    fig, ax = plt.subplots(1)
    
    # Display the image
    ax.imshow(image)
    
    # Get the Region of the Image
    each_regions = image_info['regions']
    for region in each_regions:
        label  = region["region_attributes"]["name"]
        x      = region["shape_attributes"]['x']
        y      = region["shape_attributes"]["y"]
        width  = region["shape_attributes"]["width"]
        height = region["shape_attributes"]["height"]
        
         # Create a rectangle patch
        rect = patches.Rectangle((x, y), width, height, linewidth=1, edgecolor='r', facecolor='none')

        # Add the rectangle to the current axes
        ax.add_patch(rect)

        # Add label text near the bounding box
        ax.text(x, y, label, color='r', fontsize=8, verticalalignment='top')
    plt.show()
        #display(Image.fromarray(draw_img))
```
